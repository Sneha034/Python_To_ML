# Class Properties vs Object Properties

# 1. Object Properties (Instance Variables)

Object properties are variables that belong to each individual object.

Each object gets its own separate copy.

They are usually created using self

```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

s1 = Student("Sneha", 90)
s2 = Student("Rahul", 85)

print(s1.name)
print(s2.name)
```

# 2. Class Properties (Class Variables)

Class properties are variables shared by all objects of the class.

Defined directly inside the class but outside methods.

```python
class Student:
    school = "ABC School"   # class property

    def __init__(self, name):
        self.name = name    # object property

s1 = Student("Sneha")
s2 = Student("Rahul")

print(s1.school)
print(s2.school)
```

| Feature                        | Object Property | Class Property               |
| ------------------------------ | --------------- | ---------------------------- |
| Belongs to                     | Object          | Class                        |
| Created using                  | `self.variable` | `variable` directly in class |
| Separate copy for each object? | Yes             | No                           |
| Shared among objects?          | No              | Yes                          |
| Memory usage                   | More            | Less                         |
| Accessed using                 | object          | class or object              |

# What is Abstract Method?

```python
# A method declared but not fully implemented.

# Child classes MUST implement it.
```

# Real Example Using Abstract Class

```python
from abc import ABC, abstractmethod
```

# Create Abstract Class

```python
class Animal(ABC):
    
    @abstractmethod
    def sound(self):
        pass
```

```python
class Dog(Animal):

    def sound(self):
        print("Bark")


class Cat(Animal):

    def sound(self):
        print("Meow")
```

```python
d = Dog()
d.sound()

c = Cat()
c.sound()
```

```python
from abc import ABC, abstractmethod


class Payment(ABC):

    @abstractmethod
    def pay(self, amount):
        pass


class UPI(Payment):

    def pay(self, amount):
        print(f"Paid {amount} using UPI")


class CreditCard(Payment):

    def pay(self, amount):
        print(f"Paid {amount} using Credit Card")


u = UPI()
u.pay(1000)

c = CreditCard()
c.pay(2000)
```

```python
class Employee:

    def __init__(self):
        self.__salary = 0


    def set_salary(self, amount):

        if amount > 0:
            self.__salary = amount
        else:
            print("Invalid salary")


    def get_salary(self):
        return self.__salary


e = Employee()

e.set_salary(50000)

print(e.get_salary())
```

# Types of Inheritance in Python

# 1. Single Inheritance

# One parent → one child

```python
class A:
    pass

class B(A):
    pass
```

# 2. Multiple Inheritance

One child inherits from multiple parents.

```python
class Father:

    def skills1(self):
        print("Driving")


class Mother:

    def skills2(self):
        print("Cooking")


class Child(Father, Mother):
    pass


c = Child()

c.skills1()
c.skills2()
```

# 3. Multilevel Inheritance

```python
class A:

    def show1(self):
        print("A")


class B(A):

    def show2(self):
        print("B")


class C(B):

    def show3(self):
        print("C")


c = C()

c.show1()
c.show2()
c.show3()
```

# 4. Hierarchical Inheritance

One parent → multiple children

```python
class Animal:

    def eat(self):
        print("Eating")


class Dog(Animal):
    pass


class Cat(Animal):
    pass
```

# 5. Hybrid Inheritance

Combination of multiple inheritance types.

# Method Overriding

Child class changes parent method.

```python
class Animal:

    def sound(self):
        print("Animal sound")


class Dog(Animal):

    def sound(self):
        print("Bark")


d = Dog()

d.sound()
```
