```python
#matplotlob OOA
#plt.xlabel('text')
#plt.ylabel('text')
#plt.legend()
import matplotlib.pyplot as plt

months=[1,2,3,4]
sales=[1000,1500,1200,1800]

plt.plot(months,sales,color='pink',linestyle='--',linewidth=2,marker='o',label='2026 sales data')
plt.xlabel('Months')

plt.ylabel('Sales per Month')

plt.title('Monthly Sales Data Report')

plt.legend(loc='upper left',fontsize=12)

plt.grid(color='gray',linestyle=':',linewidth=1)

plt.xlim(1,4)

plt.ylim(0,2000)

plt.xticks([1,2,3,4],['M1','M2','M3','M4'])

# plt.yticks([],[])

plt.show()
```

```python
import matplotlib.pyplot as plt

regions = ['North', 'South', 'East', 'West']

revenue = [3000, 2000, 1500, 1000]

plt.pie(
    revenue,
    labels=regions,
    autopct='%1.1f%%',
    colors=['gold', 'skyblue', 'lightgreen', 'coral']
)

plt.show()
```

```python
#plt.hist(data,bins=no_of_bins,color)
scores=[45,67,89,79,80,45,67,39,89,45,80]

plt.hist(scores,bins=5,color='pink',edgecolor='blue')

plt.xlabel('scores')

plt.title('students marks')

plt.show()
```

```python
#scatter plots

plt.scatter([1,2,3],[50,55,60],color='blue',label='class A')
plt.scatter([1,2,3],[45,50,52],color='orange',label='class B')


plt.xlabel('hours studied')

plt.ylabel('exam score')

plt.title('copasion of two classes')

plt.legend()

plt.grid(True)

plt.show()


```

```python

#subplots adn layout adjutsmensts
#linechart-sales
#barchart(expenses)


x=[1,2,3,4]
y=[10,20,15,25]

plt.subplot(1,2,1)

plt.plot(x,y)

plt.subplot(1,2,2)

plt.bar(x,y)

plt.title('bar chart')

plt.show()
```

```python
x=[1,2,3,4]
y=[10,20,15,25]

fig,ax=plt.subplots(1,2,figsize=(10,5))

ax[0].plot(x,y)
ax[0].set_title('line chart')

ax[1].bar(x,y)
ax[1].set_title('bar chart')

plt.tight_layout()

plt.show()

```

```python
x=[1,2,3,4]
y=[10,20,15,25]

fig,ax=plt.subplots(1,2,figsize=(10,5))

ax[0].plot(x,y,color='blue')
ax[0].set_title('line chart')

ax[1].bar(x,y,color='green')
ax[1].set_title('bar chart')


fig.suptitle('comparisosn of line and bar charts')
plt.tight_layout()

plt.show()

```

```python
#saving figures

#savefig

import matplotlib.pyplot as plt

x=[1,2,3,4]
y=[10,15,20,25]

plt.plot(x,y,color='blue',marker='o')
plt.title('simple line plot')

plt.xlabel('x_axis')
plt.ylabel('y_axis')

plt.savefig('line_plot.png',dpi=300,bbox_inches='tight')
plt.show()
```
