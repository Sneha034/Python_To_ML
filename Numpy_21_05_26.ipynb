```python
#advanced topics

#add
#remove
#stack
#split


"""
np.inset(array,index,value,axis=None)
array-original array
index-
value-
axis
axis =-,row wise
1 columns wise
"""

import numpy as np
arr=np.array([10,20,30,40,50,60])
print(arr)

new_arr=np.insert(arr,2,100,axis=0)

print(new_arr)


#insert in 2d array

arr_2d=np.array([[1,2],[3,4]])

new_arr_2d=np.insert(arr_2d,1,[5,6],axis=0)
print(new_arr_2d)


#append last

arr=np.array([10,20,30,40,50,60])

print(arr)

new_arr=np.append(arr,[40,50,60])

print(new_arr)



#concatenate
""" np.concatenate(array1,array2),axis=0

axis 0->vertical

axis 1>horizontal
"""

import numpy as np

arr1=np.array([1,2,3])
arr2=np.array([4,5,6])

new_arr=np.concatenate((arr1,arr2))
print(new_arr)



#removing elemnt
""" np.delete(array,index,axis=none)
flatten array
"""

arr=np.array([10,20,30,40,50,60])
print(arr)

new_arr=np.delete(arr,0)
print(new_arr)




#2d array -remove

import numpy as np
arr_2d=np.array([[1,2,3],[4,5,6]])

new_arr_2d=np.delete(arr_2d,0,axis=0)
print(new_arr_2d)
```

```python
#nan -not a number

arr1=np.array([1,2,np.nan,4,np.nan,6])

print(arr1)

print(np.isnan(arr1))


#print(np.nan == np.nan) #no we can not compare directly

#isnan
#np.nan_to_num(array,nan=value) default -0

arr=np.array([1,2,np.nan,4,np.nan,6])

cleaned_arr=np.nan_to_num(arr,nan=100)

print(cleaned_arr)


#infinite

#np.isinf(arr) 10*100

#1/0

arr2=np.array([1,2,np.inf,4,-np.inf,6])

print(np.isinf(arr2))



#replace

cleaned_arr2=np.nan_to_num(arr2,posinf=1000,neginf=1000)

print(cleaned_arr2)



```
