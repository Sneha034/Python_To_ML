```python
from IPython.utils.encoding import DEFAULT_ENCODING
#sorting

import pandas as pd

#sorting data in one column

#sort_values()

#df.sort_values(by="col_name",True/False,inplace=True)-->true-asc and desc(false)





data={
    "name":['arun','tarun','karun'],
    "age":[34,40,18],
    "salary":[1000,2000,3000]

}
df=pd.DataFrame(data)
df.sort_values(by="age",ascending=True,inplace=True)
print("sorted age")
print(df)




#
```

```python

#mul sorting
#df.sort_values(bye=["age","salary"],ascening=True,inplace=True)
data={
    "name":['arun','tarun','karun'],
    "age":[34,40,18],
    "salary":[1000,2000,3000]

}
df=pd.DataFrame(data)
df.sort_values(by=["age","salary"],ascending=[True,False],inplace=True)
print("sorted age")
print(df)

```

```python
#grouping values->groupby

#df.groupby

data={
    "name":['arun','tarun','karun'],
    "age":[34,40,18],
    "salary":[10000,50000,30000]

}
df=pd.DataFrame(data)
df.sort_values(by="age",ascending=True,inplace=True)
print("sorted age")
print(df)



print('grouped by groupby')
grouped=df.groupby("age")["salary"].sum()

print(grouped)




#groupbby on multiple columns
```
