# List Comprehension

```python
#syntax-newlist = [expression for item in iterable if condition == True]
```

# List comprehension offers a shorter syntax when you want to create a new list based on the values of an existing list.

```python
fruits = ["apple", "banana", "cherry", "kiwi", "mango"]
newlist = []

for x in fruits:
  if "a" in x:
    newlist.append(x)

print(newlist)
```

```python
fruits = ["apple", "banana", "cherry", "kiwi", "mango"]

newlist = [x for x in fruits if "a" in x]

print(newlist)
```

# Condition
-->The condition is like a filter that only accepts the items that evaluate to True.

```python
newlist = [x for x in fruits if x != "apple"]
```

# Iterable

```python
newlist = [x for x in range(10) if x < 5]
```

# Python - Sort Lists

```python
# Sort List Alphanumerically

lst1= ["orange", "mango", "kiwi", "pineapple", "banana"]

lst1.sort()

print(lst1)

#Sort the list numerically:

lst2=[100, 50, 65, 82, 23]

lst2.sort()

print("sorted by numercially:",lst2)

```

```python
# Sort Descending

lst1= ["orange", "mango", "kiwi", "pineapple", "banana"]

lst1.sort(reverse=True)

print(lst1)



lst2=[100, 50, 65, 82, 23]

lst2.sort(reverse=True)

print(lst2)
```

```python
# # | Method      | Description                                                     | Example                                      | Output      |
# | ----------- | --------------------------------------------------------------- | -------------------------------------------- | ----------- |
# | `append()`  | Adds an element at the end of the list                          | `nums = [1,2]`<br>`nums.append(3)`           | `[1,2,3]`   |
# | `clear()`   | Removes all elements from the list                              | `nums = [1,2,3]`<br>`nums.clear()`           | `[]`        |
# | `copy()`    | Returns a copy of the list                                      | `nums = [1,2,3]`<br>`new_nums = nums.copy()` | `[1,2,3]`   |
# | `count()`   | Returns the number of elements with the specified value         | `nums = [1,2,2,3]`<br>`nums.count(2)`        | `2`         |
# | `extend()`  | Adds elements of another iterable to the end of the list        | `nums = [1,2]`<br>`nums.extend([3,4])`       | `[1,2,3,4]` |
# | `index()`   | Returns the index of the first element with the specified value | `nums = [10,20,30]`<br>`nums.index(20)`      | `1`         |
# | `insert()`  | Adds an element at the specified position                       | `nums = [1,3]`<br>`nums.insert(1,2)`         | `[1,2,3]`   |
# | `pop()`     | Removes and returns the element at the specified position       | `nums = [1,2,3]`<br>`nums.pop()`             | `3`         |
# | `remove()`  | Removes the first item with the specified value                 | `nums = [1,2,3]`<br>`nums.remove(2)`         | `[1,3]`     |
# | `reverse()` | Reverses the order of the list                                  | `nums = [1,2,3]`<br>`nums.reverse()`         | `[3,2,1]`   |
# | `sort()`    | Sorts the list in ascending order                               | `nums = [3,1,2]`<br>`nums.sort()`            | `[1,2,3]`   |
```
